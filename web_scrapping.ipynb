```python
def product(product1, product2, product3):
    data=product2,product3,product1
    for text in data:
        li = text
        products, price, oldprice = li
        productloop=[product.text for product in products]
        priceloop=[prices.text for prices in price]
        oldpriceloop=[prices.text for prices in oldprice]
        data={'Product Name':productloop, 'Price':priceloop,'Old Price':oldpriceloop, 'Ratings':ratingsloop}
        df=pd.DataFrame(data,columns=['Product Name','Price','oldprice'])
        
    return df
    #print(productloop, priceloop,oldprice)
```
