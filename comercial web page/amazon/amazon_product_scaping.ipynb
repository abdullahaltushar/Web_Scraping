```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=iphones"
```

```python
wanted_list=["₹49,900","Apple iPhone 11 (64GB) - White","67,367"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_votf':'Title','rule_23zv':'Price'})
scraper.keep_rules(['rule_votf','rule_23zv'])
scraper.save('amazon-search')
```

```python
results = scraper.get_result_similar("https://www.amazon.in/s?k=oneplus+mobiles&crid=1P3IZIMYQUHIK&sprefix=oneplus+%2Caps%2C680&ref=nb_sb_ss_ts-doa-p_7_8", group_by_alias=True)
```

```python
results
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in results.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('mi_Data.csv')
```
