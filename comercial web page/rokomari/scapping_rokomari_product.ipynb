```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
web="https://www.rokomari.com/book/author/1/humayun-ahmed?ref=mm_p0"
```

```python
page=requests.get(web)
```

```python
text=page.content
text
```

```python
soup = BeautifulSoup(text, "html.parser")
soup
```

```python
products=soup.find_all('p',attrs={'class':'book-title'})
author=soup.find_all('p',attrs={'class':'book-author'})
price=soup.find_all('strike',attrs={'class':'original-price'})
#ratings=soup.find_all('span',attrs={'class':'c3XbGJ'})
```

```python
print(price)
```

```python
productloop=[product.text for product in products]
authorloop=[authors.text for authors in author ]
priceloop=[prices.text for prices in price]
#ratingsloop=[rating.text for rating in ratings]
```

```python
print(productloop)
```

```python
data={'Product Name':productloop, 'Price':priceloop,'Author':authorloop}
```
