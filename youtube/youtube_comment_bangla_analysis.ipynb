```python
punctuations=list(string.punctuation)
```

```python
data1.comment = data1.comment.apply(lambda x: " ".join(x for x in x.split() if x not in punctuations))
```

```python
import re
def replace_new_line(text):
    return re.sub(r'(\n)', ' ', text)
```

```python
data1['comment']= data1['comment'].apply(lambda x: replace_new_line(x))
```

```python
def replace_under_line(text):
    return re.sub(r'_', '', text)
```

```python
data1['comment']= data1['comment'].apply(lambda x: replace_under_line(x))
```

```python
def replace_line(text):
    return re.sub(r'|', '', text)
```

```python
data1['comment']= data1['comment'].apply(lambda x: replace_line(x))
```

```python
from bnlp import NLTKTokenizer
def tokenize(text):   
    bnltk = NLTKTokenizer()
#text = "ржЖржорж┐ ржнрж╛ржд ржЦрж╛ржЗред рж╕рзЗ ржмрж╛ржЬрж╛рж░рзЗ ржпрж╛рзЯред рждрж┐ржирж┐ ржХрж┐ рж╕рждрзНржпрж┐ржЗ ржнрж╛рж▓рзЛ ржорж╛ржирзБрж╖?"

    sentence_tokens = [bnltk.sentence_tokenize(text) for word in text]
    return sentence_tokens
```

```python
data1['sentence_token']= data1['comment'].apply(lambda x: tokenize(x)) 
```

```python
def tokenize_word(text):   
    bnltk = NLTKTokenizer()
#text = "ржЖржорж┐ ржнрж╛ржд ржЦрж╛ржЗред рж╕рзЗ ржмрж╛ржЬрж╛рж░рзЗ ржпрж╛рзЯред рждрж┐ржирж┐ ржХрж┐ рж╕рждрзНржпрж┐ржЗ ржнрж╛рж▓рзЛ ржорж╛ржирзБрж╖?"
    word_tokens = [bnltk.word_tokenize(text) for word in text]
    return  word_tokens 
```

```python
data1['word_tokens']= data1['comment'].apply(lambda x: tokenize_word(x)) 
```

```python
from bnlp.corpus import stopwords
from bnlp.corpus.util import remove_stopwords
def stopWord_remover(text):
    result = remove_stopwords(text, stopwords)
    return result
```

```python
data1['stop_word_remove']= data1['comment'].apply(lambda x: stopWord_remover(x)) 
data1
```

```python
data1['stop_word_remove'] = data1.stop_word_remove.apply(lambda x: ' '.join(x))
```
