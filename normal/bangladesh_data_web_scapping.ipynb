```python
def remove_html_tag(text):
    soup=BeautifulSoup(text,"html.parser")
    return soup.get_text()
```

```python
import re
def remove_backet(text):
     return re.sub('\[[^]]*\]', '', text)
```

```python
import string

def remove_punctuation(text):
    new_word=[]
    for word in text:
        if word not in string.punctuation:
            new_word.append(word)
    return new_word
```

```python
def puncuation(text):
    return re.sub('r[^\w\s]','',text)
```

```python
def remove_new_line(text):
    return re.sub(r'(\n)',' ',text)
```

```python
def extra_space(text):
    return re.sub(r' +',' ',text)
```

```python
import contractions
def contraction(text):
    return contractions.fix(text)
```

```python
def process(text):
    data1=remove_html_tag(text)
    data2=remove_backet(data1)
    data3=puncuation(data2)
    data4=remove_new_line(data3)
    data5= extra_space(data4)
    data6= contraction(data5)
    return data6
```

```python
data1= process(data)
data1
```

```python
import wordninja
def word_split(text):
    return wordninja.split(text)
```

```python
data=word_split(data1)
data
```

```python
import re, string, unicodedata
def remove_non_ascii(text):
    new_words = []
    for word in text:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words
```

```python
data= remove_non_ascii(data)
data
```

```python
def to_lowercase(text):
    new_word=[]
    for word in text:
        new_words= word.lower()
        new_word.append(new_words)
    return new_word
```

```python
data=to_lowercase(data)
data
```

```python
from nltk.corpus import stopwords

def stop_word_remover(text):
    new_word=[]
    for word in text:
        if word not in stopwords.words('english'):
            new_word.append(word)
    return new_word
```

```python
data= stop_word_remover(data)
data
```

```python
from nltk.stem import WordNetLemmatizer
def lammitizer(text):
    lammitizer=WordNetLemmatizer()
    lammas=[]
    for word in text:
        lamma=lammitizer.lemmatize(word)
        lammas.append(lamma)
    return lammas
```

```python
data=lammitizer(data)
data
```
