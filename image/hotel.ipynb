```python
import string
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree
```

```python
data["Name"]=data["Name"].apply(lambda x: remove_punctuation(x))
data
```

```python
import re
def replace_space(text):
    rmv_space=re.sub(r'\nHotel\n',' ',text)
    return rmv_space
```

```python
data["Name"]=data["Name"].apply(lambda x: replace_space(x))
data
```

```python
import re
def space(text):
    rmv_space=re.sub(r'\n',' ',text)
    return rmv_space
```

```python
data["Name"]=data["Name"].apply(lambda x: space(x))
data
```

```python
data["Address"]=data["Address"].apply(lambda x: remove_punctuation(x))
data["Address"]=data["Address"].apply(lambda x: space(x))
data
```

```python
data["Hotel"]=data["Hotel"].apply(lambda x: remove_punctuation(x))
data["Hotel"]=data["Hotel"].apply(lambda x: space(x))
data
```

```python
data["Price"]=data["Price"].apply(lambda x: remove_punctuation(x))
data["Price"]=data["Price"].apply(lambda x: space(x))
data
```
